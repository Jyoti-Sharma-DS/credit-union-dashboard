## OVERVIEW

**Executive Summary: Agent Performance Analysis**
<br>
This analysis evaluates employee performance using task-level data from a 3-month period, covering over 261,369 records completed by 15 agents. The objective is to uncover performance patterns, operational trends, and data-driven recommendations for improving task efficiency and agent effectiveness.

**Key Highlights:**

***Workload Distribution:***
Identified disparities in task volume across agents and task categories, with "Research" being the most common task type.

***Performance Metrics:***
Computed agent-level KPIs such as:

- Average tasks per day
- Average handle time
- Task success rate
- Issue frequency

***Trend Analysis:***
Time-series analysis revealed daily workload fluctuations with a stable long-term trend around ~3000 tasks/day.

***Correlation Insights:***
Clear relationships between:

- Agents and task specialization
- Task types and successful outcomes

***Clustering & Segmentation:***
Applied KMeans clustering to segment agents into:

- High performers
- Balanced performers
- Agents needing support

**Outcome:**
This analysis enables data-backed decisions around task allocation, performance coaching, and process improvement to optimize both individual and team efficiency.

### Understand the key metrics such as

```python
# Average tasks per task
print( "i.   On an average {} tasks are being handled per day .".\
      format(np.round(Pd_data['TASK_COUNT'].sum()/ Pd_data['DATE'].nunique())))

# Average Handle time per task
print("ii.  On an average agents took {} seconds to complete a task".format(\
      np.round(Pd_data['HANDLE_TIME'].sum()/ Pd_data['TASK_COUNT'].sum(), 2)))

print("iii. Success ratio for all tasks performed is {}.".format(\
      np.round(sum(Pd_data['Outcome_Category'] == 'Successful Completion')/len(Pd_data), 2)))
```

#### Generate the Summary by agent

```python
# Calculate total active days per agent
active_days = Pd_data.groupby('AGENT_ID')['DATE'].nunique()
issue_categories = ['Contact Issues', 'Employer/Employee Issues']

# Group by Agent_ID to create features
agent_features = Pd_data.groupby('AGENT_ID').agg(
    Total_Tasks=('TASK_COUNT', 'sum'),
    Avg_Handle_Time=('HANDLE_TIME', 'mean'),
    Max_Handle_Time=('HANDLE_TIME', 'max'),
    Min_Handle_Time=('HANDLE_TIME', 'min'),
    Median_Handle_Time=('HANDLE_TIME', 'median'),
    Task_Variance=('HANDLE_TIME', 'std'),
    Unique_Tasks=('Task_Category', 'nunique'),
    Success_Rate=('Outcome_Category', lambda x: (x=='Successful Completion').mean()),
    Tasks_With_Issues=('Outcome_Category', lambda x: x.isin(issue_categories).mean())
).fillna(0)

# Add Average Tasks per Day
agent_features['Avg_Tasks_Per_Day'] = agent_features['Total_Tasks'] / active_days

# Fill any potential NaN (due to division by zero if any agent has no active days)
agent_features['Avg_Tasks_Per_Day'] = agent_features['Avg_Tasks_Per_Day'].fillna(0)

agent_features = agent_features.round({
    'Avg_Handle_Time': 2, 
    'Task_Variance': 2,
    'Success_Rate': 2, 
    'Avg_Tasks_Per_Day': 2
})
```

**Agent-Level Insights** <br>
Let's understand if there are statistically significant differences in Avg_Handle_Time, Success_Rate, or Avg_Tasks_Per_Day across agents.  


```python
plt.figure(figsize=(16, 9))
sns.boxplot(data=Pd_data, x='AGENT_ID', y='HANDLE_TIME')
plt.xticks(rotation=45)
plt.title('Handle Time Distribution by Agent')
plt.ylabel('Handle Time (seconds)')
plt.xlabel('Agent')
plt.tight_layout()
plt.show()
```

```python
# Prepare handle time values per agent
anova_data = [Pd_data[Pd_data['AGENT_ID'] == agent]['HANDLE_TIME'].values 
              for agent in Pd_data['AGENT_ID'].unique()]

f_stat, p_value = f_oneway(*anova_data)
print(f"\n ANOVA - Avg Handle Time Differences Across Agents")
print(f"F-statistic: {f_stat:.2f} | p-value: {p_value:.4f}")
if p_value < 0.05:
    print(" Significant differences detected between agents.")
else:
    print(" No statistically significant difference in handle times.")
```

Since ANOVA is showing statistical difference, let's perform Tukey's HSD to understand which agents differ from each other (~ In terms of handling time)

```python
# Prepare data for Tukey HSD
tukey_data = Pd_data[['HANDLE_TIME', 'AGENT_ID']].copy()
tukey_data = tukey_data.dropna()

# Run Tukey HSD
tukey_result = pairwise_tukeyhsd(endog=tukey_data['HANDLE_TIME'], groups=tukey_data['AGENT_ID'], alpha=0.05)

print("\n Tukey HSD Test Results:")
print(tukey_result)
```

```python
 
# Step 1: Initialize the matrix
agents = list(tukey_result.groupsunique)
sig_matrix = pd.DataFrame(False, index=agents, columns=agents)

# Step 2: Fill the matrix with Tukey HSD significance results
for i in range(len(tukey_result.reject)):
    g1 = tukey_result._results_table.data[i+1][0]  # agent1
    g2 = tukey_result._results_table.data[i+1][1]  # agent2
    is_significant = tukey_result.reject[i]
    if is_significant:
        sig_matrix.loc[g1, g2] = True
        sig_matrix.loc[g2, g1] = True  # symmetric matrix

# Step 3: Plot the heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(sig_matrix, cmap="Reds", cbar=False, linewidths=0.5, annot=True, fmt="")
plt.title(' Tukey HSD: Significant Differences in Handle Time Between Agents')
plt.xlabel('Agent')
plt.ylabel('Agent')
plt.xticks(rotation=45)
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
# Select features for clustering
features = [ 'Total_Tasks','Avg_Tasks_Per_Day', 'Avg_Handle_Time', 'Task_Variance', 
            'Unique_Tasks', 'Success_Rate', 'Tasks_With_Issues']

# ensure no NA values
X = agent_features[features].fillna(0)

# Scale the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

#### Identify the optimal(#) of clusters

```python
inertia = []
# Check for clusters
k_range = range(1, 6)  # check quickly for k=1 to k=5
for k in k_range:
    km = KMeans(n_clusters=k, random_state=42)
    km.fit(X_scaled)
    inertia.append(km.inertia_)

plt.plot(k_range, inertia, 'bo-')
plt.xlabel('Number of Clusters')
plt.ylabel('Inertia')
plt.title('Elbow Method for Optimal Clusters')
plt.show()
```

We have identified 3 as the optimal number of cluster based on the elbow method.

```python
# Use k=3 as a reasonable starting point
kmeans = KMeans(n_clusters=3, random_state=42)
agent_features['Cluster'] = kmeans.fit_predict(X_scaled)

# Check results quickly
print(display(agent_features[['Total_Tasks', 'Avg_Handle_Time', 'Success_Rate','Tasks_With_Issues', 'Cluster']]))
```

```python
pca = PCA(n_components=2)
principal_components = pca.fit_transform(X_scaled)

agent_features['PCA1'] = principal_components[:, 0]
agent_features['PCA2'] = principal_components[:, 1] 
```

```python

plt.figure(figsize=(8, 6))

# Plot PCA scatter with clusters
sns.scatterplot(x='PCA1', y='PCA2', data=agent_features, hue='Cluster', palette='viridis', s=100)

# Add Agent ID as text labels (assuming index contains agent IDs)
for i in range(agent_features.shape[0]):
    plt.text(
        agent_features['PCA1'][i] + 0.05,  # small offset to avoid overlapping with point
        agent_features['PCA2'][i],
        agent_features.index[i],
        fontsize=8
    )

plt.title('Agent Clusters Based on Performance')
plt.xlabel('PCA1')
plt.ylabel('PCA2')
plt.tight_layout()
plt.show()
```
